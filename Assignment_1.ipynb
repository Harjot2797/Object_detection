```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Dataset Creation
data = {
    "Patient ID": [101, 102, 103, 104, 105],
    "Age": [45, 33, 29, 56, 33],
    "Diagnosis": ["Pneumonia", "Fracture", "Diabetes", "Hypertension", "Pneumonia"],
    "Length of Stay (Days)": [5, 3, 7, 4, 6],
    "Hospital Department": ["General Medicine", "Orthopedics", "Endocrinology", "Cardiology", "General Medicine"]
}

# Load dataset into a DataFrame
df = pd.DataFrame(data)

# Step 1: Data Cleaning
# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values (Example: Add some NaNs to demonstrate)
df.loc[1, "Diagnosis"] = np.nan
df.loc[3, "Length of Stay (Days)"] = np.nan
df["Diagnosis"] = df["Diagnosis"].fillna("Unknown")  # Fill missing Diagnosis with 'Unknown'
df["Length of Stay (Days)"] = df["Length of Stay (Days)"].fillna(df["Length of Stay (Days)"].mean())  # Fill missing LOS with mean

# Step 2: Basic Statistical Analysis
mean_age = df["Age"].mean()
median_los = df["Length of Stay (Days)"].median()
std_dev_age = df["Age"].std()

print("Basic Statistical Analysis:")
print(f"Mean Age: {mean_age:.2f}")
print(f"Median Length of Stay: {median_los}")
print(f"Standard Deviation of Age: {std_dev_age:.2f}")

# Step 3: Visualization
# Bar Chart: Average Length of Stay by Department
los_by_department = df.groupby("Hospital Department")["Length of Stay (Days)"].mean()
los_by_department.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Average Length of Stay by Department")
plt.ylabel("Average Length of Stay (Days)")
plt.xlabel("Hospital Department")
plt.show()

# Pie Chart: Age Group Distribution
df["Age Group"] = pd.cut(df["Age"], bins=[18, 30, 50, np.inf], labels=["18-30", "31-50", "51+"])
age_group_distribution = df["Age Group"].value_counts()
age_group_distribution.plot(kind="pie", autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Age Group Distribution")
plt.ylabel("")  # Remove y-axis label for cleaner visualization
plt.show()

# Line Chart: Trends in Length of Stay vs. Age
sns.lineplot(data=df, x="Age", y="Length of Stay (Days)", marker="o", color="orange")
plt.title("Trends in Length of Stay vs. Age")
plt.xlabel("Age")
plt.ylabel("Length of Stay (Days)")
plt.show()
```

```python
import pandas as pd
import numpy as np
# Sample Dataset Creation (with Missing and Inaccurate Data)
data = {
    "Patient ID": [101, 102, 103, 104, 105],
    "Age": [45, np.nan, 29, -56, 33],  # -56 is inaccurate
    "Diagnosis": ["Pneumonia", None, "Diabetes", "Hypertension", ""],  # Empty and None values
    "Length of Stay (Days)": [5, 3, np.nan, 4, 6],
    "Hospital Department": ["General Medicine", "Orthopedics", None, "Cardiology", "General Medicine"]
}
# Load dataset into a DataFrame
df = pd.DataFrame(data)
# Step 1: Identify Missing or Inaccurate Data
print("Dataset with Missing and Inaccurate Data:")
print(df)

# Step 2: Handle Missing Values
# Replace missing 'Diagnosis' with 'Unknown'
df["Diagnosis"] = df["Diagnosis"].fillna("Unknown")

# Replace missing 'Length of Stay (Days)' with mean
df["Length of Stay (Days)"] = df["Length of Stay (Days)"].fillna(df["Length of Stay (Days)"].mean())

# Replace missing 'Hospital Department' with 'Unknown'
df["Hospital Department"] = df["Hospital Department"].fillna("Unknown")

# Step 3: Handle Inaccurate Data
# Remove rows with negative or zero Age
df = df[df["Age"] > 0]

# Step 4: Standardize Formats
# Trim extra spaces in string columns
df["Diagnosis"] = df["Diagnosis"].str.strip()
df["Hospital Department"] = df["Hospital Department"].str.strip()

# Final Cleaned Dataset
print("\nCleaned Dataset:")
print(df)

# Summary of Steps Taken
summary = {
    "Missing Values Replaced": {
        "Diagnosis": "Filled with 'Unknown'",
        "Length of Stay": "Filled with mean",
        "Hospital Department": "Filled with 'Unknown'"
    },
    "Inaccurate Data Removed": "Rows with negative or zero Age",
    "Standardization": "String trimming for columns"
}

print("\nSummary of Data Cleaning Steps:")
for key, value in summary.items():
    print(f"{key}: {value}")

```
